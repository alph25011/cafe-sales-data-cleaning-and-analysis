```python
import pandas as pd
```

```python
df=pd.read_csv("Cleaned_cafe_sales.csv")
```

```python
#===========================================================================
#HAVE RECRUITED SOME INSIGHTS
#===========================================================================
```

**1) Top 3 items that are sold most**

```python
df["item"].value_counts().head(3)
```

**2) Revenue per Item sold**

```python
df.groupby("item")["total_spent"].sum().sort_values(ascending=False)
```

**3) Average spending per order**

```python
print(df["total_spent"].mean().round(2))
```

**4) Quantity Sold per item**

```python
df.groupby("item")["quantity"].sum().sort_values(ascending=False)
```
